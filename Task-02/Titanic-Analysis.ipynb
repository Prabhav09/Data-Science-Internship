```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\prabh\Downloads\Titanic.csv")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df = df.drop(columns = ["Cabin"])
```

```python
df.isnull().sum()
```

### Handling Missing Values
Missing data inspection showed gaps in the Age, Cabin, Embarked Column. The Age column contained a moderate number of missing values, so median imputation was applied to preserve distribution stability. The Cabin column had a very high proportion of missing entries & was removed because it would not provide reliable analytical value. The Embarked column had olny a few missing values, which were was filled using the most frequent category.

```python
sns.barplot(x ="Sex", y ="Survived", data = df)
plt.title("Survival Rate by Gender")
plt.show()
```

### Who Survived more: males or females?
The grouped analysis shows that female passengers have a higher survival rate than male passengers.Therefore, females survived more compared to males

```python
sns.barplot(x="Pclass", y="Survived", data = df)
plt.title("Survival Rate by Class")
plt.show()
```

### Did Passenger Class Affect Survival Chances?
The survival rate varies across passenger classes. First-Class passenger show the highest survival rate, followed by second class, While third class has the lowest survival rate, This indicates that passenger class significantly affected survival chances.

```python
plt.hist(df["Age"], bins = 20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### What was the survival rate by age group?
After categorizing passenger into age groups, the computed survival rate show variation across groups. Younger passengers tend to have higher survival rates,While older passengers generally show lower survival probability.
